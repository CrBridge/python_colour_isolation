```python
import skimage
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = skimage.io.imread("Homer.jpeg")
```

```python
plt.imshow(image)
```

```python
homer_gray = skimage.color.rgb2gray(image)
homer_gray = np.stack((homer_gray,) * 3, axis=2)
plt.imshow(homer_gray, cmap="gray")
```

```python
homer_hue = skimage.color.rgb2hsv(image)
plt.imshow(homer_hue)
homer_hue = homer_hue[:, :, 0]
```

```python
lower_hue = 270/360
upper_hue = 330/360
mask = (homer_hue >= lower_hue) & (homer_hue <= upper_hue)
```

```python
img = np.copy(homer_gray)
img[mask] = image[mask] / 255.0
```

```python
plt.imshow(img)
```
